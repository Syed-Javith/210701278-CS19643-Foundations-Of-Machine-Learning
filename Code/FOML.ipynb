<h1>Build Model

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,RandomForestClassifier,BaggingClassifier,StackingClassifier,VotingClassifier,ExtraTreesClassifier
```

```python
acc_list=[]
model_list=[]
trained_list=[]
def acc_calc(y_test,y_pred,model,trained_model):
    acc=accuracy_score(y_test,y_pred)*100
    acc_list.append(acc)
    trained_list.append(trained_model)
    model_list.append(model)
    print(acc)
```

```python
df.head(1)
```

```python
X = df.drop(['year','churn'],axis=1)
y = df['churn']
```

```python
X.head(1)
```

```python
pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X,y=smote.fit_resample(X,y)
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train.shape
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
y_pred=log_reg.predict(X_test)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
acc_calc(y_test,y_pred,'Logistic Regression',log_reg)
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(X_train,y_train)
y_pred=random_forest.predict(X_test)
acc_calc(y_test,y_pred,'Random Forest',random_forest)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train)
y_pred=decision_tree.predict(X_test)
acc_calc(y_test,y_pred,'Desicion Tree',decision_tree)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
svc = SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
acc_calc(y_test,y_pred,'Support Vector machine',svc)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
acc_calc(y_test,y_pred,'KNN',knn)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
gaussian_naive = GaussianNB()
gaussian_naive.fit(X_train,y_train)
y_pred=gaussian_naive.predict(X_test)
acc_calc(y_test,y_pred,'Gaussian NB',gaussian_naive)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
# multinomial_nb = MultinomialNB()
# multinomial_nb.fit(X_train,y_train)
# y_pred=multinomial_nb.predict(X_test)
# acc_calc(y_test,y_pred,'Multinomial NB',multinomial_nb)
# sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
# print(classification_report(y_test,y_pred))
```

```python
bernoulli = BernoulliNB()
bernoulli.fit(X_train,y_train)
y_pred=bernoulli.predict(X_test)
acc_calc(y_test,y_pred,'Bernoulli NB',bernoulli)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_pred=gb.predict(X_test)
acc_calc(y_test,y_pred,'Gradient Boosting',gb)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
ab = AdaBoostClassifier()
ab.fit(X_train,y_train)
y_pred=ab.predict(X_test)
acc_calc(y_test,y_pred,'Adaptive Boosting',ab)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
bagging = BaggingClassifier(estimator=random_forest,n_estimators=10)
bagging.fit(X_train,y_train)
y_pred=bagging.predict(X_test)
acc_calc(y_test,y_pred,'Bagging',bagging)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
stack=StackingClassifier(estimators =[('KNN',knn), ('NB',gaussian_naive)], final_estimator = log_reg)
stack.fit(X_train,y_train)
y_pred=stack.predict(X_test)
acc_calc(y_test,y_pred,'Stacking',stack)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
# voting=VotingClassifier(estimators =[('KNN',knn), ('NB',gaussian_naive)], final_estimator = log_reg)
# voting.fit(X_train,y_train)
# y_pred=voting.predict(X_test)
# acc_calc(y_test,y_pred,'Voting')
# sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
# print(classification_report(y_test,y_pred))
```

```python
extra=ExtraTreesClassifier()
extra.fit(X_train,y_train)
y_pred=extra.predict(X_test)
acc_calc(y_test,y_pred,'Extra Trees',extra)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
gbk = GradientBoostingClassifier()
ne = np.arange(1,20)
dep = np.arange(1,10)
param_grid = {'n_estimators' : ne,'max_depth' : dep}
gbk_cv = GridSearchCV(gbk, param_grid=param_grid, cv=5)
gbk_cv.fit(X, y)
y_pred=gbk_cv.predict(X_test)
acc_calc(y_test,y_pred,'Gradient Boosting with Hyperparameter tuning',gbk_cv)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
ne = np.arange(1,20)
dep = np.arange(1,15)
param_grid = {'n_estimators' : ne,'max_depth' : dep}
gbk_cv = GridSearchCV(random_forest, param_grid=param_grid, cv=5)
gbk_cv.fit(X, y)
y_pred=gbk_cv.predict(X_test)
acc_calc(y_test,y_pred,'Random Forest with Hyperparameter tuning',gbk_cv)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
dep = np.arange(1,10)
param_grid = {'max_depth' : dep}
gbk_cv = GridSearchCV(decision_tree, param_grid=param_grid, cv=5)
gbk_cv.fit(X, y)
y_pred=gbk_cv.predict(X_test)
acc_calc(y_test,y_pred,'Decision Tree with Hyperparameter tuning',gbk_cv)
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print(classification_report(y_test,y_pred))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
nnn = Sequential()
nnn.add(Dense(8,activation='relu'))
Dropout(0.3)
nnn.add(Dense(16,activation='relu'))
Dropout(0.3)
nnn.add(Dense(32,activation='relu'))
Dropout(0.3)
nnn.add(Dense(64,activation='relu'))
nnn.add(Dense(1,activation='sigmoid'))

nnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=10)
# early_stop_acc = EarlyStopping(monitor='accuracy',mode='max',verbose=1,patience=10)
```

```python
nnn.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,callbacks=[early_stop])
```

```python
history = pd.DataFrame(nnn.history.history)
```
